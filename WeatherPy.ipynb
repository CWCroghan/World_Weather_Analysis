```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
for coordinate in coordinates:
    print(coordinate[0],coordinate[1])
```

```python
# Create list to hold cities information
cities = []

#Identify the nearest city for each random lat & long combination
for coor in coordinates:
    city = citipy.nearest_city(coor[0], coor[1]).city_name
    
    #Add to cities only if it is unique
    if city not in cities:
        cities.append(city)

#Checking to see how many unique cities we found
len(cities)
```
